```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better graphs
sns.set(style="whitegrid")


# Load the Titanic dataset
df = pd.read_csv('train.csv')

# Show the first 5 rows
df.head()

# Get basic info about the dataset
df.info()

# Get summary statistics for numerical columns
df.describe()

# Check for missing values
df.isnull().sum()


# Fill missing values for 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop the 'Cabin' column due to excessive missing values
df.drop('Cabin', axis=1, inplace=True)

# Fill missing values for 'Embarked' with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Check if there are any remaining missing values
df.isnull().sum()

# Countplot for Survived column
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()


# Histogram for Age column
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Age')
plt.show()

# Boxplot for Age vs Survived
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()

# Histogram for Fare column
sns.histplot(df['Fare'], kde=True, bins=20)
plt.title('Distribution of Fare')
plt.show()

# Compute the correlation matrix
corr = df.corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()







```

Summary of Findings
Survival Distribution:

The countplot of Survived indicates an imbalance in the dataset, with more non-survivors (0) than survivors (1). This is typical for the Titanic disaster, where the majority of passengers did not survive. The survival rate is significantly low compared to the number of non-survivors.

Age Distribution:

The histogram of Age shows that most passengers on the Titanic were young adults, with a noticeable concentration of passengers aged between 20 and 40. The distribution is slightly skewed to the right, with fewer passengers in the older age groups. There is a small peak around children and infants as well.

This suggests that the age range of passengers was fairly broad but skewed towards younger individuals, possibly due to more families traveling on the ship.

Age vs Survived:

The boxplot comparing Age and Survived indicates that the passengers who survived tend to have a slightly lower median age than those who did not. However, the age distribution for both survivors and non-survivors is fairly wide, suggesting that survival was not strictly dependent on age. There is also a wider spread of ages among non-survivors, which might indicate that survival was influenced by factors other than just age (e.g., class, gender).

Fare Distribution:

The histogram of Fare reveals that the majority of passengers paid low fares, with a few passengers paying significantly higher fares. The distribution is right-skewed, suggesting that most people traveled in the lower-price brackets, while a small number of first-class passengers paid much higher fares. This might reflect the class structure of the Titanic, where wealthy passengers in first class paid much higher fares than those in second or third class.

Correlation:

The correlation matrix heatmap shows that Pclass and Fare are negatively correlated, meaning that passengers in lower classes (3rd class) tended to pay lower fares. Age doesnt seem to have a strong correlation with other features except for Pclass.

Interestingly, Survived has a moderate positive correlation with Pclass (those in 1st and 2nd class had higher survival rates), which aligns with historical data where wealthier passengers had better access to lifeboats and other survival resources.
